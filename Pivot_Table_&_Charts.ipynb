#Pivot Table & Charts

**Question 1: What is the purpose of a PivotTable in Excel?**

 - A PivotTable in Excel is used to summarize, analyze, and present large datasets in a simple and meaningful way without changing the original data.

Purpose of a PivotTable:

To organize data by categories such as Region, Product, or Date

To calculate totals, sums, averages, and counts

To compare performance across regions or products

To make data analysis quick and easy

**Question 4. Explain the difference between Sorting and Filtering with an example from this datast.**
 - Difference between Sorting and Filtering

 **Sorting**

Sorting is used to rearrange data in a specific order (ascending or descending) based on a column.

Example :
If we sort the Sales column from Highest to Lowest, the record with ₹497 (South – Oranges) will appear at the top, and ₹53 (West – Oranges) will appear at the bottom.

**Filtering**

Filtering is used to display only specific data while hiding the rest temporarily.

Example :
If we apply a filter on Region = West, Excel will show only West region sales records and hide South, East, and North data.

Difference Summary
| Sorting           | Filtering                   |
| ----------------- | --------------------------- |
| Rearranges data   | Hides unwanted data         |
| Shows all records | Shows selected records only |
| Changes row order | Keeps original order        |

**Question 9. Group dates in the PivotTable into Months and calculate the average sales per month.**

 - Group Dates into Months & Calculate Average Sales per Month.

Step 1: Create the PivotTable

Select the entire dataset

Go to Insert → PivotTable

Choose New Worksheet

Drag:

Date → Rows

Sales → Values

Step 2: Group Dates by Month

Right-click on any date in the PivotTable

**Question 10. Combine Apples and Oranges into one group called Fruits and compare its sales with other products.**

- To analyze combined sales of Apples + Oranges as one category (Fruits) and compare it with other products (Mangoes).

Step 1: Create a PivotTable

Select the entire dataset

Go to Insert → PivotTable

Choose New Worksheet

Drag:

Product → Rows

Sales → Values (Sum of Sales)

Step 2: Group Apples and Oranges

In the PivotTable, hold Ctrl

Select Apples and Oranges

Right-click → Group

Rename Group1 to Fruits

Now Apples + Oranges = Fruits

Step 3: Compare Sales (Based on Your Data)

**PivotTable Result**
| Product Group                 | Total Sales (₹) |
| ----------------------------- | --------------- |
| **Fruits (Apples + Oranges)** | **4,308**       |
| **Mangoes**                   | **2,521**       |

Thus, by grouping Apples and Oranges into a single category called Fruits, it is observed that the Fruits category generates higher total sales compared to other products, indicating it is the major contributor to overall revenue.
